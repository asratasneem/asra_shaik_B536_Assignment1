```python
# Function to count 3-letter words (trinucleotides) in the genome sequence
def count_trinucleotides(file_path):
    # Open the file and read all lines
    file = open(file_path, 'r')
    lines = file.readlines()
    file.close()

    # Combine all lines into a single string, except the first line
    sequence = ''
    for line in lines[1:]:  # Skip the first line (sequence name)
        sequence += line.strip()  # Remove newline characters

    # Dictionary to store counts of each 3-letter word
    counts = {}

    # Loop through the sequence to find every 3-letter combination
    for i in range(len(sequence) - 2):
        # Extract a 3-letter word starting at position i
        trinucleotide = sequence[i:i + 3]

        # If the 3-letter word is already counted, increase its count
        if trinucleotide in counts:
            counts[trinucleotide] += 1
        # If it's not counted yet, add it with a count of 1
        else:
            counts[trinucleotide] = 1

    # Print the counts of each 3-letter word in alphabetical order
    for trinucleotide in sorted(counts):
        print(trinucleotide, counts[trinucleotide])



# Provide the path to the genome file (e.g., 'sequence.fasta')
file_path = 'sequence.fasta'
count_trinucleotides(file_path)

```
